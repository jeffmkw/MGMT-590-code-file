```python
import urllib2
```

```python
#Q1
#URL looking for:
#http://chart.finance.yahoo.com/table.csv?s=AAPL&a=0&b=1&c=2003&d=11&e=31&f=2003&g=m&ignore=.csv
company = 'AAPL'
url = 'http://chart.finance.yahoo.com/table.csv?s='+company+'&a=0&b=1&c=2003&d=11&e=31&f=2003&g=m&ignore=.csv'
url
```

```python
csv = urllib2.urlopen(url).read()
csv
```

```python
rows = csv.splitlines(1)
rows
```

```python
clean = ''
for i in rows:
    a = i.split(',')
    #print i
    newrow =a[0]+','+a[5]+','+a[6]
    clean += newrow
clean
```

```python
#write file
outfile = open('C:\\Users\\Jeffery\\Google Drive\\Homework\\web data 590\\fileoutput\\HW3Q1.csv','w')
outfile.write(clean)
outfile.close()
print "Q1 ends here"
```

```python
#Q2 starts
data = ''
company_list = ['AAPL','F','MSFT']

for company in company_list:
    url = 'http://chart.finance.yahoo.com/table.csv?s='+company+'&a=0&b=1&c=2003&d=11&e=31&f=2003&g=m&ignore=.csv'
    csv = urllib2.urlopen(url).read()
    rows = csv.splitlines(1)
    if company == company_list[0]:
            data += 'ticker,'+rows[0]
    for i in rows[1:len(rows)]:
        data += company +','+i

print data
```

```python
#split the new data
new_data = ''
temp = data.splitlines(1)
for a in temp:
    i = a.split(',')
    new_data += i[0] +','+ i[1] +','+ i[6] +','+ i[7]
new_data
```

```python
#write file
outfile = open('C:\\Users\\Jeffery\\Google Drive\\Homework\\web data 590\\fileoutput\\HW3Q2.csv','w')
outfile.write(new_data)
outfile.close()
print "Q2 ends here"
```
