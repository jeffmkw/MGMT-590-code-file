```python
o_ign = pd.read_csv('C:\Users\Jeffery\Google Drive\Homework\web data 590\project\ign.csv')
print(list(o_ign.columns.values))
o_ign = o_ign.rename(columns={'title':'Name','release_year':'Year','platform':'Platform'})
o_ign.head(5)
```

```python
o_sales = pd.read_csv('C:\Users\Jeffery\Google Drive\Homework\web data 590\project\\vgsales.csv')
o_sales.head(5)
```

```python
set(o_sales['Platform'])
```

```python
print(o_ign.columns.values)
print(o_sales.columns.values)
```

```python
ign = o_ign.loc[o_ign['Year']>2013]
print(len(ign))
sales = o_sales.loc[o_sales['Year']>2013]
print(len(sales))
```

```python
dataset = pd.merge(ign, sales, on=['Name', 'Platform'], how='inner')
print(len(dataset))
```

```python
dataset =  dataset.drop(['Unnamed: 0','release_day','Rank','Year_y'],axis = 1)
dataset.rename(columns={'Year_x':"year"})
```
